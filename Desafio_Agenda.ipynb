# **Desafio Agendas**




```python
# importando as bibliotecas necessárias
import numpy as np
import datetime as dt
from datetime import timedelta
import pandas as pd
```

```python
#Acabei copiando do documento mas acredito que esses dados
#serão inseridos por algum tipo de sistema
inicioA = ['9:00', '20:00']
pessoaA = [['9:00', '10:30'], 
           ['12:00', '13:00'], 
           ['16:00', '18:00']]

inicioB = ['10:00', '18:30']
pessoaB = [['10:00', '11:30'], 
           ['12:30', '14:30'], 
           ['14:30', '15:00'], 
           ['16:00','17:00']]
```

```python
stt = dt.timedelta(hours=9)
stp = dt.timedelta(minutes=30)
end = dt.timedelta(hours=20)

entireRange = 2*((end-stt).seconds//3600)+1

aa = pd.to_timedelta(stt+np.arange(entireRange)*stp, unit='H')
data = {'Grade Horaria' : aa}
df = pd.DataFrame(data)

df
```

```python
dfA = pd.DataFrame(pessoaA,columns = ["Inicio", "Fim"])
dfB = pd.DataFrame(pessoaB,columns = ["Inicio", "Fim"])
dfA
```

```python
dfB
```

```python
#Criando as colunas relativas a pessoa A e B em df
df["Pessoa_A"] = "Livre"
df["Pessoa_B"] = "Livre"
df
```

```python
#Agora que tenho o DF com todos os Horários livre, preciso 
# marcar como ocupado os horarios em dfA e dfB. 

nLinesdfA = dfA['Inicio'].count()
nLinesdfB = dfB['Inicio'].count()

for i in np.arange(nLinesdfA) : 
  df.Pessoa_A[ (df['Grade Horaria'] >= '0 days '+dfA['Inicio'].iloc[i]+':00') & 
      (df['Grade Horaria'] <= '0 days '+dfA['Fim'].iloc[i]+':00')] = "Ocupado"


for i in np.arange(nLinesdfB) : 
  df.Pessoa_B[ (df['Grade Horaria'] >= '0 days '+dfB['Inicio'].iloc[i]+':00') & 
      (df['Grade Horaria'] <= '0 days '+dfB['Fim'].iloc[i]+':00')] = "Ocupado"
```
